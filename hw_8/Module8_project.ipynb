## Работа с API

### Получить данные обо всех книгах

```python
api_base_url = 'https://www.anapioficeandfire.com/api'
```

```python
books_list = requests.get(
    f'{api_base_url}/books?page=1&pageSize=20'
).json()
```

```python
len(books_list)
```

```python
df_books = pd.DataFrame(books_list)
df_books.head()
```

### Получить данные обо всех домах Вестероса

```python
houses_list = []
pagesize = 50

for page in range(1,50):
    response = requests.get(f'{api_base_url}/houses?page={page}&pageSize={pagesize}').json()
    if not response:
        break
    houses_list.extend(response)
```

```python
len(houses_list)
```

```python
df_houses = pd.DataFrame(houses_list)
df_houses.head()
```

```python
df_houses.shape
```

```python
# в дтатафрейме есть пустые значения, обозначенные [] (в json были пустым списком), заменим их на привычный NaN
df_houses_clean = df_houses.map(lambda x: np.nan if x == [] else x) # map - применяет функцию поэлементно (apply - для строк/столбцов)
df_houses_clean.head(4)
```

```python
df_houses_clean.isna().sum()
```

### Получить данные обо всех домах Вестероса, у которых есть девиз

```python
houses_list_with_titles = []
pagesize = 50
parametrs = {'hasTitles' : True} 

for page in range(1,50):
    response_houses = requests.get(
        f'{api_base_url}/houses?page={page}&pageSize={pagesize}',
        params = parametrs
    ).json()
    if not response_houses:
        break
    houses_list_with_titles.extend(response_houses)
```

```python
len(houses_list_with_titles)
```

```python
df_houses_with_titles = pd.DataFrame(houses_list_with_titles)
df_houses_with_titles_clean = df_houses_with_titles.map(lambda x: np.nan if x == [] else x)
df_houses_with_titles_clean.head()
```

```python
df_houses_with_titles.shape
```

```python
df_houses_with_titles_clean.isna().sum()
```
